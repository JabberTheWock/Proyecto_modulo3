## *1.3 Modelo que representa el problema*
\begin{equation}
F = m_{1} \frac{d^2x_{1}}{dt^2} = -k_{1}x_{1}+k_{2}(x_{2}-x_{1})
\end{equation}

\begin{equation}
F = m_{2} \frac{d^2x_{2}}{dt^2} = -k_{2}(x_{2}-x_{1})
\end{equation}

### *Enunciado que representa el problema*

Se tienen 2 masas de 1 slug cada una conectadas mediante dos resortes, el resorte 1 tiene una constante ($k_1$) de 6lb/ft y el resorte 2 tiene una constante ($k_2$) de 4lb/ft. El movimiento de las masas comienza desde la posición de equilibrio.
La velocidad inicial de la masa 1 ($m_1$) es de 1ft/s y la de la masa 2 ($m_2$) es de -1ft/s. Dar las ecuaciones de movimiento de cada masa y graficar.



## *1.4 Solución*

```python
#importamos las librerías que se van a ocupar
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import sympy as sym
from scipy.integrate import odeint
```

```python
# Definimos una función que nos genere la ecuación diferencial a resolver
def ec_diferencial(x,t):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    x4 = x[3]
    k1 = 6
    k2 = 4
    m1 = 1
    m2 = 1
    #aquí ponemos los valores que va a tomar cada derivada de las variables
    dx1 = x3
    dx2 = x4
    dx3 = ((-k1/m1)*x1)+((k2/m1)*(x2-x1))
    dx4 = ((-k2/m2)*(x2-x1))
    return [dx1,dx2,dx3,dx4]
    
```

```python
#De acuerdo a las ecuaciones que representan al problema se definen con las variables que ya definimos
#Se definen las funciones para las ecuaciones de movimiento de las masas
#hmin = 1e-5
t = np.linspace(0,20,2000)
x0 = np.array([0,0,1,-1])

x= odeint(ec_diferencial,x0,t)
```

## *1.5 Visualización de la solución del problema*

```python
#Graficando la solución numérica de x1
plt.figure(figsize = (8,6))
plt.plot(t,x[:,0]);
```

```python
#Graficando la solución numérica de x2
plt.figure(figsize = (8,6))
plt.plot(t,x[:,1]);
```

```python
#Graficando la solución numérica de x3, velocidad
plt.figure(figsize = (8,6))
plt.plot(t,x[:,2]);
```

```python
#Graficando la solución numérica de x4, velocidad
plt.figure(figsize = (8,6))
plt.plot(t,x[:,3]);
```

Nuestra solución analítica es ...

$x_1(t)=\frac{6}{5}\frac{\sin\sqrt12t}{\sqrt12}-\frac{1}{5}\frac{\sin\sqrt2t}{\sqrt2}$

$x_2(t)=\frac{-3}{5}\frac{\sin\sqrt12t}{\sqrt12}-\frac{2}{5}\frac{\sin\sqrt2t}{\sqrt2}$

Ahora vamos a comparar la solución numérica que nos da python con la solución analítica

```python
#Definimos las soluciones analíticas

x1_analítica = ((6/5)*((np.sin(np.sqrt(12)*t))/(np.sqrt(12))))-((1/5)*((np.sin(np.sqrt(2)*t))/np.sqrt(2)))
x2_analítica = ((-3/5)*((np.sin(np.sqrt(12)*t))/np.sqrt(12)))-((2/5)*((np.sin(np.sqrt(2)*t))/(np.sqrt(2))))
```

```python
#Gráfica de solución analítica de x1
plt.figure(figsize = (8,6))
plt.plot(t,x1_analítica);
```

```python
#Gráfica de solución analítica de x2
plt.figure(figsize = (8,6))
plt.plot(t,x2_analítica);
```

```python
#Gráfico comparativo de las soluciones analíticas y numéricas para x1
plt.figure(figsize = (8,6))

plt.plot(t,x[:,0],'b-',linewidth=2,label='Solución numérica de x1')
plt.plot(t,x1_analítica,'m-',linewidth=2,label='Solución analítica de x1')

plt.xlabel('x')
plt.ylabel('y')

plt.grid()
plt.legend(loc='best')
plt.show();
```

```python
#Gráfico comparativo de las soluciones analíticas y numéricas para x2
plt.figure(figsize = (8,6))

plt.plot(t,x[:,1],'g-',linewidth=2,label='Solución numérica de x2')
plt.plot(t,x2_analítica,'c-',linewidth=2,label='Solución analítica de x2')

plt.xlabel('x')
plt.ylabel('y')

plt.grid()
plt.legend(loc='best')
plt.show();
```
